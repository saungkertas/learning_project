# GeoPandas MultiPoint Example\n\nCreate GeoDataFrame → MultiPoint → GeoJSON.

```python
import geopandas as gpd\nfrom shapely.geometry import Point, MultiPoint\nimport json
```

```python
# Sample points\npoints = [Point(0, 0), Point(1, 1), Point(2, 0)]\n\n# GeoDataFrame\ngdf = gpd.GeoDataFrame(geometry=points, crs='EPSG:4326')\nprint(gdf)\ngdf.plot()
```

```python
# Combine to MultiPoint\nmultipoint = MultiPoint(points)\nprint(multipoint)\n\n# New GDF with MultiPoint\ngdf_mp = gpd.GeoDataFrame([{'name': 'Sample MultiPoint'}], geometry=[multipoint], crs='EPSG:4326')\ngdf_mp.plot()
```

```python
# Export GeoJSON\ngdf_mp.to_file('multipoint.geojson', driver='GeoJSON')\nprint('Exported multipoint.geojson')
```
