# Buffer Analysis: Subway Service Areas (Day 3)\n\n500m buffers around stations (project to UTM → buffer → back to WGS84).

```python
import geopandas as gpd\nfrom shapely.geometry import Point\nimport matplotlib.pyplot as plt
```

```python
# Sample subway stations (NYC-ish lon/lat)\nstations_data = [\n    {'name': 'Station A', 'coords': (-74.0, 40.7)},\n    {'name': 'Station B', 'coords': (-73.9, 40.75)},\n    {'name': 'Station C', 'coords': (-74.1, 40.72)},\n]\n\npoints = [Point(lon, lat) for d in stations_data for lon, lat in [d['coords']]]\nstations_gdf = gpd.GeoDataFrame(stations_data, geometry=points, crs='EPSG:4326')\nstations_gdf
```

```python
# Project to UTM Zone 18N (meters, NYC area)\nutm_crs = 'EPSG:32618'\nstations_utm = stations_gdf.to_crs(utm_crs)\nprint('Original CRS:', stations_gdf.crs)\nprint('UTM CRS:', stations_utm.crs)
```

```python
# 500m buffer (in UTM meters)\nbuffers_utm = stations_utm.copy()\nbuffers_utm['geometry'] = stations_utm.geometry.buffer(500)\nbuffers_utm
```

```python
# Project back to WGS84\nbuffers_wgs84 = buffers_utm.to_crs('EPSG:4326')\nbuffers_wgs84
```

```python
# Visualize\nfig, ax = plt.subplots(figsize=(10,8))\nstations_gdf.plot(ax=ax, color='red', markersize=100, label='Stations')\nbuffers_wgs84.plot(ax=ax, alpha=0.5, edgecolor='blue', facecolor='lightblue', label='500m Buffers')\nax.set_title('Subway Station 500m Buffers')\nax.legend()\nplt.show()
```

```python
# Export GeoJSON\nbuffers_wgs84.to_file('subway_buffers.geojson', driver='GeoJSON')\nprint('Saved subway_buffers.geojson – view on geojson.io')
```
