# Network Analysis: Shortest Path (Day 5)\n\nOSMnx street graph → networkx shortest route.

```python
import osmnx as ox\nimport networkx as nx\nimport geopandas as gpd\nfrom shapely.geometry import LineString\nimport folium
```

```python
# NYC box (drivable graph)\nplace = 'New York, NY, USA'\nG = ox.graph_from_place(place, network_type='drive')\nprint('Graph nodes/edges:', G.number_of_nodes(), '/', G.number_of_edges())
```

```python
# Start/end points (NYC lat/lon)\nstart = (40.7589, -73.9851)  # Times Square\nend = (40.6892, -74.0445)  # Battery Park\n\n# Nearest nodes\nstart_node = ox.distance.nearest_nodes(G, start[1], start[0])\nend_node = ox.distance.nearest_nodes(G, end[1], end[0])\nprint('Start node:', start_node, 'End node:', end_node)
```

```python
# Shortest path (distance)\nroute = nx.shortest_path(G, start_node, end_node, weight='length')\nprint('Route nodes:', len(route))\n\n# Route geometry\nroute_coords = [(G.nodes[i]['y'], G.nodes[i]['x']) for i in route]\nroute_line = LineString(route_coords)\nprint(route_line)
```

```python
# Folium map with route\nm = folium.Map(location=[40.75, -74.0], zoom_start=12)\nfolium.GeoJson(route_line).add_to(m)\nm.save('shortest_route.html')\nm
```

```python
# Export GeoJSON\nroute_gdf = gpd.GeoDataFrame(geometry=[route_line], crs='EPSG:4326')\nroute_gdf.to_file('shortest_route.geojson', driver='GeoJSON')\nprint('Saved shortest_route.geojson')
```

## Notes\n- Open shortest_route.html for interactive map.\n- Use ox.distance.shortest_path for more options.\n- Larger area? Filter network_type='walk'.
